+ - Suma
- - Resta
* - Multiplicacion
/ - Divisiom
// - Division sin decimales
% - Resto (modulo)

```python
producto_1 = "papas fritas"
costo_producto1 = 15000
producto_2 = "Niko"
costo_producto2 = 6000
capital = 50000
dolar = 6500

# Realizamos la suma
costo_total = costo_producto1 + costo_producto2
print (costo_total)

# Realizamos la resta
vuelto = capital - costo_total
print(vuelto)
# Realizamos la division
vuelto_dolar = vuelto//dolar
print(vuelto_dolar)
```
