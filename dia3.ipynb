```python
palabra = input("Ingrese una palabra: ")
letra = input("Ingrese la letra que desea buscar: ")

def comparar(palabra, letra):
    for caracter in palabra:
        if caracter == letra:
         return [letra]
    return letra

print(comparar(palabra, letra))
```
