Challenge 1 ü§∫ 10m
Escribir un programa que cree un diccionario vac√≠o y lo vaya llenado con informaci√≥n sobre una persona (por ejemplo nombre, edad, sexo, tel√©fono, correo electr√≥nico, etc.) que se le pida al usuario. 

Cada vez que se a√±ada un nuevo dato debe imprimirse el contenido del diccionario.

```python
nombre = input("Ingresa el nombre")
apellido = input ("Ingrese el apellido")
edad = int(input("Ingrese su edad"))
sexo = input("ingrese el dato")
telefono = int(input ("Ingrese su numero de telefono"))

datos = {"Nombre":nombre, "Apellido":apellido, "Edad":edad, "Sexo":sexo, "Telefono": telefono}
print(datos)



```

```python
# 1. Creamos el diccionario vac√≠o
persona = {}

# Variable de control para el bucle
continuar = True

print("--- Registro de Informaci√≥n ---")

while continuar:
    # 2. Pedimos al usuario qu√© tipo de dato quiere guardar (la clave)
    clave = input("\n¬øQu√© dato quieres introducir? (ej: nombre, edad, tlf): ").lower()
    
    # 3. Pedimos el valor del dato
    valor = input(f"Introduce el/la {clave}: ")
    
    # 4. Guardamos el dato en el diccionario
    persona[clave] = valor
    
    # 5. Imprimimos el contenido actual del diccionario
    print("Contenido actual del diccionario:")
    print(persona)
    
    # 6. Preguntamos si quiere seguir
    respuesta = input("\n¬øQuieres a√±adir m√°s informaci√≥n? (si/no): ").lower()
    if respuesta == "no":
        continuar = False

print("\n--- Proceso finalizado ---")
print("Diccionario final:", persona)
```

```python
persona = {}
continuar = True

print("--- Registro de Informaci√≥n ---")

while continuar:
    print("\n¬øQu√© quieres hacer?")
    print("1. A√±adir contacto")
    print("2. Modificar contacto")
    print("3. Eliminar contacto")
    print("4. Ver datos")
    print("5. Salir")

    opcion = input("Elige una opci√≥n (1-5): ")

    if opcion == "1":
        clave = input("Ingrese el nombre del contacto ").lower()
        valor = input(f"Introduce el/la {clave}: ")
        persona[clave] = valor
        print("contacto a√±adido correctamente.")

    elif opcion == "2":
        clave = input("¬øQu√© contacto quieres modificar?: ").lower()
        if clave in persona:
            nuevo_valor = input(f"Introduce el nuevo contacto para {clave}: ")
            persona[clave] = nuevo_valor
            print("contacto modificado correctamente.")
        else:
            print("Ese contacto no existe.")

    elif opcion == "3":
        clave = input("¬øQu√© contacto quieres eliminar?: ").lower()
        if clave in persona:
            del persona[clave]
            print("contacto eliminado correctamente.")
        else:
            print("Ese contacto no existe.")

    elif opcion == "4":
        print("\nCcontactos actuales:")
        print(persona)

    elif opcion == "5":
        continuar = False

    else:
        print("Opci√≥n no v√°lida. Int√©ntalo de nuevo.")
        

print("\n--- Proceso finalizado ---")
print("contactos finales:", persona)
```
