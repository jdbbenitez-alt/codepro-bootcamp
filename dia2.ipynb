```python
#Usamos condicionales + operaciones de comparacion

edad= 18

# Primera condicion
if edad < 18:
    print("Es menor de edad")
# Segunda condicion
else: 
    print("Es mayor de edad")
```

```python
# Si tenemos mas condiciones
edad_laboral =17

if edad_laboral <=18:
    print("menor de edad")
elif edad_laboral >65:
    print("Jubilado")
else:
    print("Activo Laboralmente")
```

```python
#Preguntar edad
edad= int(input("Cuantos años tienes"))

#Condiciones
if edad <18:
    print("Juguito nomas")
elif edad >=18 and edad <21:
    print("Podes tomar pero no comprar")
elif edad >21:
    print("Hace lo que quieras")
       
 
```

```python
edad_laboral = 18
trabaja = False

#Primera condicio 

if edad_laboral<18:
    print("Estudiante, menor de edad")
elif edad_laboral <=65 and trabaja == True:
    print("activo laboralmente")
#Terceracondicion
elif edad_laboral <65 and trabaja == False:
    print("Estudiante")
#Cuarta y ultima condicion
else:
    print("Jubilado")
```
